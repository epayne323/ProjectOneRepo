```python
import pandas as pd
import os
```

```python
file = pd.read_csv(os.path.join("..","Data","Raw","us_wineries_2009_2019.csv"),header=None)
print(file)
```

```python
producing_wineries_df = pd.DataFrame(file)
producing_wineries_df
```
