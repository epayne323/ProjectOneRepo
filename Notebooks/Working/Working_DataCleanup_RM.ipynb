```python
liver = pd.read_csv(os.path.join("..","Data","Raw","Chronic_Liver_Disease_Cirrhosis_Mortality_by_State.csv"),header=0)
print(liver)
```

```python
liver_df = pd.DataFrame(liver)
liver_df.head()
```

```python
liver_df.count()
```

```python
liver_df = liver_df.drop("URL", axis=1)
liver_df.head()
```

```python
yearmin = liver_df["YEAR"].min()
yearmax = liver_df["YEAR"].max()
totalyears = yearmax - yearmin
print(yearmin)
print(yearmax)
print(totalyears)
```

```python
liver_state_death = liver_df.groupby("STATE").sum()
liver_death_df = pd.DataFrame(liver_state_death)
liver_death_df = liver_death_df.drop(["YEAR", "RATE"],axis=1)
liver_death_df.rename(columns = {'DEATHS' : 'TOTAL DEATHS'}, inplace = True)
liver_death_df.head()
```

```python
liver_state_rate = liver_df.groupby("STATE").mean()
liver_rate_df = pd.DataFrame(liver_state_rate)
liver_rate_df = liver_rate_df.drop(["YEAR", "DEATHS"],axis=1)
liver_rate_df.rename(columns = {'RATE' : 'AVERAGE DEATH RATE'}, inplace = True)
liver_rate_df.head()
```

```python
liver_overall_df = liver_rate_df.merge(liver_death_df, on = "STATE", how = 'outer')
liver_overall_df.count()
```

```python
liver_overall_df.head()
```

```python
liver_state_year = liver_df.groupby(["YEAR","STATE"]).sum().reset_index()
liver_state_year.drop(liver_state_year[liver_state_year["YEAR"]>yearmin].index, inplace = True)
liver_state_year.rename(columns = {'RATE' : '2005 DEATH RATE', 'DEATHS': '2005 DEATH COUNT'}, inplace = True)
liver_state_year = liver_state_year.drop("YEAR",axis=1)
liver_overall_df = liver_overall_df.merge(liver_state_year, on = "STATE", how = 'outer')
liver_overall_df.head()
```

```python
liver_overall_df.count()
```

```python
liver_state_year = liver_df.groupby(["YEAR","STATE"]).sum().reset_index()
liver_state_year.drop(liver_state_year[liver_state_year["YEAR"]!=2014].index, inplace = True)
liver_state_year.rename(columns = {'RATE' : '2014 DEATH RATE', 'DEATHS': '2014 DEATH COUNT'}, inplace = True)
liver_state_year = liver_state_year.drop("YEAR",axis=1)
liver_overall_df = liver_overall_df.merge(liver_state_year, on = "STATE", how = 'outer')
liver_overall_df.count()
```

```python
liver_state_year = liver_df.groupby(["YEAR","STATE"]).sum().reset_index()
liver_state_year.drop(liver_state_year[liver_state_year["YEAR"]!=2015].index, inplace = True)
liver_state_year.rename(columns = {'RATE' : '2015 DEATH RATE', 'DEATHS': '2015 DEATH COUNT'}, inplace = True)
liver_state_year = liver_state_year.drop("YEAR",axis=1)
liver_overall_df = liver_overall_df.merge(liver_state_year, on = "STATE", how = 'outer')
liver_overall_df.head()
```

```python
liver_state_year = liver_df.groupby(["YEAR","STATE"]).sum().reset_index()
liver_state_year.drop(liver_state_year[liver_state_year["YEAR"]!=2016].index, inplace = True)
liver_state_year.rename(columns = {'RATE' : '2016 DEATH RATE', 'DEATHS': '2016 DEATH COUNT'}, inplace = True)
liver_state_year = liver_state_year.drop("YEAR",axis=1)
liver_overall_df = liver_overall_df.merge(liver_state_year, on = "STATE", how = 'outer')
liver_overall_df.head()
```

```python
liver_overall_df.count()
```

```python
liver_state_endyear = liver_df.groupby(["YEAR","STATE"]).sum().reset_index()
liver_state_endyear.drop(liver_state_endyear[liver_state_endyear["YEAR"]<yearmax].index, inplace = True)
liver_state_endyear.rename(columns = {'RATE' : '2017 DEATH RATE', 'DEATHS': '2017 DEATH COUNT'}, inplace = True)
liver_state_endyear = liver_state_endyear.drop("YEAR",axis=1)
liver_state_endyear.head()
```

```python
liver_overall_df = liver_overall_df.merge(liver_state_endyear, on = "STATE", how = 'outer')
liver_overall_df.count()
```

```python
liver_overall_df.head()
```

```python
liver_overall_df['14 - 17 DEATH RATE CHANGE'] = liver_overall_df['2017 DEATH RATE'] - liver_overall_df['2014 DEATH RATE']
liver_overall_df['14 - 17 DEATH RATE % CHANGE'] = (liver_overall_df['2017 DEATH RATE'] / liver_overall_df['2014 DEATH RATE']) - 1
liver_overall_df['14 - 17 DEATHS % CHANGE'] = (liver_overall_df['2017 DEATH COUNT'] / liver_overall_df['2014 DEATH COUNT']) - 1
liver_overall_df
```

```python
liver_overall_df.to_csv(os.path.join("..","Data","Clean","Liver_Disease_Cirrhosis_Mortality_Clean.csv"), index = False)
```
